```python
df2=df1.drop(df1[["area_type","availability","society","balcony"]],axis=1)
df2.head()
```

```python
df3=df2.dropna()
df3.head()
```

```python
df3['size'].unique()
```

```python
df3["BHK"]=df3['size'].apply(lambda x:int(x.split(' ')[0]))
df3.head()                            
```

```python
df4=df3.drop(df3[['size']],axis=1)
df4.head()
```

```python
df4['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True;
    
```

```python
(df4[~df4["total_sqft"].apply(is_float)]).total_sqft.unique()
```

```python
def transform(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4['total_sqft']=df4["total_sqft"].apply(transform)
df4.head()
```

```python
(df4[~df4["total_sqft"].apply(is_float)])
```

```python
df4["price/sqft"]=(df4['price']*100000)/df4['total_sqft']
df4.head()
```

```python
len(df4.location.unique())
```

```python
df4.location=df4.location.apply(lambda x:x.strip())
location_count=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_count
```

```python
less_count_location=location_count[location_count<=10]
less_count_location
```

```python
df4.location=df4.location.apply(lambda x:'other'if x in less_count_location else x)
len(df4.location.unique())
```

```python
df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
df4[:11]
```

```python
df4.shape
```

```python
df4.head()
```

```python
df5=df4[~((df4.total_sqft/df4.BHK)<300)]
df5.shape
```

```python
df6=df5[df5.bath<df5.BHK+2]
df6.shape
```

```python
def remove_price_outliers(df):
    df_output=pd.DataFrame()
    for key,location_info in df.groupby('location'):
        mean=np.mean(location_info['price/sqft'])
        std=np.std(location_info['price/sqft'])
        reduce_df=location_info[(location_info['price/sqft']>(mean-std)) & (location_info['price/sqft']<=(mean+std))]
        df_output=pd.concat([df_output,reduce_df],ignore_index=True)
    return df_output

df7=remove_price_outliers(df6)
df7.shape
```

```python
def remove_location_price_outliers(df):
    exclude_idx=np.array([])
    for location,location_info in df.groupby('location'):
        bhk_info={}
        for bhk,bhk_stat in location_info.groupby('BHK'):
             bhk_info[bhk]={
            'mean':np.mean(bhk_stat['price/sqft']),
            'std':np.std(bhk_stat['price/sqft']),
            'count':bhk_stat.shape[0]
        }
        for bhk,bhk_stat in location_info.groupby('BHK'):
            prev_info=bhk_info.get(bhk-1)
            if prev_info and prev_info['count']>5:
                exclude_idx=np.append(exclude_idx,bhk_stat[bhk_stat['price/sqft']<(prev_info['mean'])].index.values)
    return df.drop(exclude_idx,axis='index')     
        
df8=remove_location_price_outliers(df7)
df8.shape 
```

```python
df9=df8.drop(['price/sqft'],axis=1)
df9.shape
```

```python
dummies=pd.get_dummies(df9.location);
df10=pd.concat([df9,dummies.drop('other',axis=1)],axis='columns')
df10.head()
```

```python
df11=df10.drop('location',axis=1)
df11.head()
```

```python
df11.shape
```

```python
X=df11.drop('price',axis=1)
X.head()
```

```python
y=df11.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, ShuffleSplit
import pandas as pd

def find_best_model(X, y):
    params_info = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                #'normalize': [True, False]
            }
        },
        'lasso_regression': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    score = []
    for model_name, config in params_info.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        score.append({
            'model': model_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(score, columns=['model', 'best_score', 'best_params'])
    
find_best_model(X,y)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
model=LinearRegression()
model.fit(X_train,y_train)

def predict_price(location,sqft,bath,bhk):
    loc_idx=np.where(X.columns==location)[0][0]
    
    x=np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=bhk
    if loc_idx>=0:
        x[loc_idx]=1
    return model.predict([x])[0]
    
```
